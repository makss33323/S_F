```python
# инициализируем информацию об одежде
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing = pd.DataFrame(clothing_list)
clothing.columns = ['size', 'type']
clothing
```

```python
# создаем объект OrdinalEncoder, col - имя столбца, mapping - словарь с описанием кодировки
ord_encoder = ce.OrdinalEncoder(mapping=[{
	'col': 'size',
	'mapping': {'xxs': 1, 'xs': 2, 's': 3, 
                'm': 4, 'l': 5, 'xxl': 6}
}])
# применяем трансформацию к столбцу
data_bin = ord_encoder.fit_transform(clothing[['size']])
# добавляем результат к исходному DataFrame
clothing = pd.concat([clothing, data_bin], axis=1)

clothing
```

```python
# Создаем свежий датасет для демонстрации OneHot кодирования
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing_onehot = pd.DataFrame(clothing_list)
clothing_onehot.columns = ['size', 'type']

print("Исходный датасет:")
print(clothing_onehot)
print(f"\nУникальные значения признака 'type': {clothing_onehot['type'].unique()}")
print(f"Количество уникальных значений: {clothing_onehot['type'].nunique()}")
```

```python
# Применяем OneHot кодирование к признаку 'type'
# Используем класс OneHotEncoder библиотеки category_encoders
# Указываем в cols наименование признака type для кодировки, иначе будут закодированы все строковые столбцы

encoder = ce.OneHotEncoder(cols=['type'], use_cat_names=True)  # указываем столбец для кодирования
type_bin = encoder.fit_transform(clothing_onehot[['type']])
clothing_onehot = pd.concat([clothing_onehot, type_bin], axis=1)

print("Результат OneHot кодирования:")
print(clothing_onehot)
```

```python
# Создаем свежий датасет для демонстрации Binary кодирования
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing_binary = pd.DataFrame(clothing_list)
clothing_binary.columns = ['size', 'type']

print("Исходный датасет:")
print(clothing_binary)
print(f"\nУникальные значения признака 'type': {clothing_binary['type'].unique()}")
print(f"Количество уникальных значений: {clothing_binary['type'].nunique()}")
```

```python
# Применяем Binary кодирование к признаку 'type'
# Используем класс BinaryEncoder библиотеки category_encoders

bin_encoder = ce.BinaryEncoder(cols=['type'])  # указываем столбец для кодирования
type_bin = bin_encoder.fit_transform(clothing_binary[['type']])
clothing_binary = pd.concat([clothing_binary, type_bin], axis=1)

print("Результат Binary кодирования:")
print(clothing_binary)
```

```python
# Сравним количество созданных признаков для разных методов кодирования

print("=== СРАВНЕНИЕ МЕТОДОВ КОДИРОВАНИЯ ===")
print(f"Исходный признак 'type' имеет {clothing_binary['type'].nunique()} уникальных значений: {list(clothing_binary['type'].unique())}")
print()

# Считаем количество новых столбцов для каждого метода
binary_cols = [col for col in clothing_binary.columns if col.startswith('type_')]
print(f"Binary кодирование создало {len(binary_cols)} новых признаков: {binary_cols}")

# Для сравнения с OneHot из предыдущего примера
onehot_cols_count = 4  # мы знаем, что OneHot создал 4 столбца
print(f"OneHot кодирование создало бы {onehot_cols_count} новых признаков")
print()

print("=== ЭФФЕКТИВНОСТЬ ===")
print(f"Экономия места: Binary использует {len(binary_cols)} столбцов вместо {onehot_cols_count}")
print(f"Сжатие: в {onehot_cols_count/len(binary_cols):.1f} раза меньше столбцов!")
```

```python
# Загружаем датасет с винами для демонстрации Binary кодирования
import pandas as pd
import category_encoders as ce
import math

# Загружаем данные
wine_df = pd.read_csv('../Data/wine_cleared.csv')

print("=== АНАЛИЗ ПРИЗНАКА COUNTRY В ДАТАСЕТЕ С ВИНАМИ ===")
print(f"Размер датасета: {wine_df.shape}")
print(f"Уникальные значения country: {wine_df['country'].nunique()}")
print(f"Пропущенные значения: {wine_df['country'].isna().sum()}")

print(f"\nТоп-10 стран по количеству вин:")
country_counts = wine_df['country'].value_counts().head(10)
for country, count in country_counts.items():
    print(f"  {country}: {count} вин")

# Показываем все уникальные страны (первые 20)
print(f"\nПервые 20 уникальных стран:")
unique_countries = sorted([c for c in wine_df['country'].unique() if pd.notna(c)])
for i, country in enumerate(unique_countries[:20]):
    print(f"{i+1:2d}. {country}")

if len(unique_countries) > 20:
    print(f"... и еще {len(unique_countries) - 20} стран")
```

```python
# Применяем Binary кодирование к признаку country
print("=== ПРИМЕНЕНИЕ BINARY КОДИРОВАНИЯ К COUNTRY ===")

# Создаем Binary кодировщик для country
binary_encoder = ce.BinaryEncoder(cols=['country'])

# Применяем кодирование
country_binary_encoded = binary_encoder.fit_transform(wine_df[['country']])

print(f"Размер исходного DataFrame: {wine_df.shape}")
print(f"Размер закодированных данных: {country_binary_encoded.shape}")

# Показываем названия новых столбцов
binary_columns = [col for col in country_binary_encoded.columns if col.startswith('country_')]
print(f"\n✅ Создано новых признаков: {len(binary_columns)}")
print(f"Названия новых признаков: {binary_columns}")

# Добавляем закодированные признаки к исходному DataFrame
wine_df_with_binary = pd.concat([wine_df, country_binary_encoded], axis=1)

print(f"\nРазмер DataFrame после добавления Binary кодирования: {wine_df_with_binary.shape}")
print(f"Добавлено столбцов: {wine_df_with_binary.shape[1] - wine_df.shape[1]}")
```

```python
# Анализ эффективности Binary кодирования на реальных данных
print("=== АНАЛИЗ ЭФФЕКТИВНОСТИ BINARY КОДИРОВАНИЯ ===")

unique_countries_count = wine_df['country'].nunique()
print(f"Количество уникальных стран: {unique_countries_count}")
print(f"Binary кодирование создало: {len(binary_columns)} признаков")
print(f"OneHot кодирование создало бы: {unique_countries_count} признаков")

efficiency = unique_countries_count / len(binary_columns)
print(f"💾 Экономия места: в {efficiency:.1f} раза меньше столбцов!")

# Формула для расчета количества битов
required_bits = math.ceil(math.log2(unique_countries_count))
print(f"\n📊 Теоретически требуется битов: ⌈log₂({unique_countries_count})⌉ = {required_bits}")
print(f"📊 Фактически создано: {len(binary_columns)} признаков")

print(f"\n=== ПРИМЕРЫ BINARY КОДИРОВАНИЯ ===")
sample_countries = ['US', 'Italy', 'France', 'Spain', 'Portugal']
for country in sample_countries:
    if country in wine_df['country'].values:
        idx = wine_df[wine_df['country'] == country].index[0]
        country_row = country_binary_encoded.loc[idx]
        
        print(f"Страна: {country}")
        binary_values = []
        for col in binary_columns:
            binary_values.append(str(country_row[col]))
        print(f"Binary код: [{', '.join(binary_values)}]")
        print("-" * 30)

print(f"\n🎯 ИТОГОВЫЙ ОТВЕТ:")
print(f"После двоичного кодирования признака 'country' образовалось {len(binary_columns)} новых признаков")
```
