```python
# –∏–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫—É –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –∫–æ–¥–∏—Ä–æ–≤—â–∏–∫–∞–º–∏
import category_encoders as ce
import pandas as pd
```

```python
# –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ–± –æ–¥–µ–∂–¥–µ
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing = pd.DataFrame(clothing_list)
clothing.columns = ['size', 'type']
clothing
```

```python
# —Å–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç OrdinalEncoder, col - –∏–º—è —Å—Ç–æ–ª–±—Ü–∞, mapping - —Å–ª–æ–≤–∞—Ä—å —Å –æ–ø–∏—Å–∞–Ω–∏–µ–º –∫–æ–¥–∏—Ä–æ–≤–∫–∏
ord_encoder = ce.OrdinalEncoder(mapping=[{
	'col': 'size',
	'mapping': {'xxs': 1, 'xs': 2, 's': 3, 
                'm': 4, 'l': 5, 'xxl': 6}
}])
# –ø—Ä–∏–º–µ–Ω—è–µ–º —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü–∏—é –∫ —Å—Ç–æ–ª–±—Ü—É
data_bin = ord_encoder.fit_transform(clothing[['size']])
# –¥–æ–±–∞–≤–ª—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∫ –∏—Å—Ö–æ–¥–Ω–æ–º—É DataFrame
clothing = pd.concat([clothing, data_bin], axis=1)

clothing
```

```python
# –∏–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫—É –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –∫–æ–¥–∏—Ä–æ–≤—â–∏–∫–∞–º–∏
import category_encoders as ce
import pandas as pd
```

```python
# –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ–± –æ–¥–µ–∂–¥–µ
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing = pd.DataFrame(clothing_list)
clothing.columns = ['size', 'type']
clothing
```

```python
# —Å–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç OrdinalEncoder, col - –∏–º—è —Å—Ç–æ–ª–±—Ü–∞, mapping - —Å–ª–æ–≤–∞—Ä—å —Å –æ–ø–∏—Å–∞–Ω–∏–µ–º –∫–æ–¥–∏—Ä–æ–≤–∫–∏
ord_encoder = ce.OrdinalEncoder(mapping=[{
	'col': 'size',
	'mapping': {'xxs': 1, 'xs': 2, 's': 3, 
                'm': 4, 'l': 5, 'xxl': 6}
}])
# –ø—Ä–∏–º–µ–Ω—è–µ–º —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü–∏—é –∫ —Å—Ç–æ–ª–±—Ü—É
data_bin = ord_encoder.fit_transform(clothing[['size']])
# –¥–æ–±–∞–≤–ª—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∫ –∏—Å—Ö–æ–¥–Ω–æ–º—É DataFrame
clothing = pd.concat([clothing, data_bin], axis=1)

clothing
```

```python
# –°–æ–∑–¥–∞–µ–º —Å–≤–µ–∂–∏–π –¥–∞—Ç–∞—Å–µ—Ç –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ OneHot –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing_onehot = pd.DataFrame(clothing_list)
clothing_onehot.columns = ['size', 'type']

print("–ò—Å—Ö–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç:")
print(clothing_onehot)
print(f"\n–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–∞ 'type': {clothing_onehot['type'].unique()}")
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: {clothing_onehot['type'].nunique()}")
```

```python
# –ü—Ä–∏–º–µ–Ω—è–µ–º OneHot –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –∫ –ø—Ä–∏–∑–Ω–∞–∫—É 'type'
# –ò—Å–ø–æ–ª—å–∑—É–µ–º –∫–ª–∞—Å—Å OneHotEncoder –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ category_encoders
# –£–∫–∞–∑—ã–≤–∞–µ–º –≤ cols –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ type –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∫–∏, –∏–Ω–∞—á–µ –±—É–¥—É—Ç –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω—ã –≤—Å–µ —Å—Ç—Ä–æ–∫–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã

encoder = ce.OneHotEncoder(cols=['type'], use_cat_names=True)  # —É–∫–∞–∑—ã–≤–∞–µ–º —Å—Ç–æ–ª–±–µ—Ü –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è
type_bin = encoder.fit_transform(clothing_onehot[['type']])
clothing_onehot = pd.concat([clothing_onehot, type_bin], axis=1)

print("–†–µ–∑—É–ª—å—Ç–∞—Ç OneHot –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è:")
print(clothing_onehot)
```

```python
# –°–æ–∑–¥–∞–µ–º —Å–≤–µ–∂–∏–π –¥–∞—Ç–∞—Å–µ—Ç –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ Binary –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing_binary = pd.DataFrame(clothing_list)
clothing_binary.columns = ['size', 'type']

print("–ò—Å—Ö–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç:")
print(clothing_binary)
print(f"\n–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–∞ 'type': {clothing_binary['type'].unique()}")
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: {clothing_binary['type'].nunique()}")
```

```python
# –ü—Ä–∏–º–µ–Ω—è–µ–º Binary –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –∫ –ø—Ä–∏–∑–Ω–∞–∫—É 'type'
# –ò—Å–ø–æ–ª—å–∑—É–µ–º –∫–ª–∞—Å—Å BinaryEncoder –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ category_encoders

bin_encoder = ce.BinaryEncoder(cols=['type'])  # —É–∫–∞–∑—ã–≤–∞–µ–º —Å—Ç–æ–ª–±–µ—Ü –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è
type_bin = bin_encoder.fit_transform(clothing_binary[['type']])
clothing_binary = pd.concat([clothing_binary, type_bin], axis=1)

print("–†–µ–∑—É–ª—å—Ç–∞—Ç Binary –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è:")
print(clothing_binary)
```

```python
# –°—Ä–∞–≤–Ω–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–∑–¥–∞–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è

print("=== –°–†–ê–í–ù–ï–ù–ò–ï –ú–ï–¢–û–î–û–í –ö–û–î–ò–†–û–í–ê–ù–ò–Ø ===")
print(f"–ò—Å—Ö–æ–¥–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫ 'type' –∏–º–µ–µ—Ç {clothing_binary['type'].nunique()} —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: {list(clothing_binary['type'].unique())}")
print()

# –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –º–µ—Ç–æ–¥–∞
binary_cols = [col for col in clothing_binary.columns if col.startswith('type_')]
print(f"Binary –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ —Å–æ–∑–¥–∞–ª–æ {len(binary_cols)} –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {binary_cols}")

# –î–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å OneHot –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—Ä–∏–º–µ—Ä–∞
onehot_cols_count = 4  # –º—ã –∑–Ω–∞–µ–º, —á—Ç–æ OneHot —Å–æ–∑–¥–∞–ª 4 —Å—Ç–æ–ª–±—Ü–∞
print(f"OneHot –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ —Å–æ–∑–¥–∞–ª–æ –±—ã {onehot_cols_count} –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")
print()

print("=== –≠–§–§–ï–ö–¢–ò–í–ù–û–°–¢–¨ ===")
print(f"–≠–∫–æ–Ω–æ–º–∏—è –º–µ—Å—Ç–∞: Binary –∏—Å–ø–æ–ª—å–∑—É–µ—Ç {len(binary_cols)} —Å—Ç–æ–ª–±—Ü–æ–≤ –≤–º–µ—Å—Ç–æ {onehot_cols_count}")
print(f"–°–∂–∞—Ç–∏–µ: –≤ {onehot_cols_count/len(binary_cols):.1f} —Ä–∞–∑–∞ –º–µ–Ω—å—à–µ —Å—Ç–æ–ª–±—Ü–æ–≤!")
```

```python
# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞—Ç–∞—Å–µ—Ç —Å –≤–∏–Ω–∞–º–∏ –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ Binary –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è
import pandas as pd
import category_encoders as ce
import math

# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
wine_df = pd.read_csv('../Data/wine_cleared.csv')

print("=== –ê–ù–ê–õ–ò–ó –ü–†–ò–ó–ù–ê–ö–ê COUNTRY –í –î–ê–¢–ê–°–ï–¢–ï –° –í–ò–ù–ê–ú–ò ===")
print(f"–†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞: {wine_df.shape}")
print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è country: {wine_df['country'].nunique()}")
print(f"–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è: {wine_df['country'].isna().sum()}")

print(f"\n–¢–æ–ø-10 —Å—Ç—Ä–∞–Ω –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –≤–∏–Ω:")
country_counts = wine_df['country'].value_counts().head(10)
for country, count in country_counts.items():
    print(f"  {country}: {count} –≤–∏–Ω")

# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –≤—Å–µ —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å—Ç—Ä–∞–Ω—ã (–ø–µ—Ä–≤—ã–µ 20)
print(f"\n–ü–µ—Ä–≤—ã–µ 20 —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å—Ç—Ä–∞–Ω:")
unique_countries = sorted([c for c in wine_df['country'].unique() if pd.notna(c)])
for i, country in enumerate(unique_countries[:20]):
    print(f"{i+1:2d}. {country}")

if len(unique_countries) > 20:
    print(f"... –∏ –µ—â–µ {len(unique_countries) - 20} —Å—Ç—Ä–∞–Ω")
```

```python
# –ü—Ä–∏–º–µ–Ω—è–µ–º Binary –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –∫ –ø—Ä–∏–∑–Ω–∞–∫—É country
print("=== –ü–†–ò–ú–ï–ù–ï–ù–ò–ï BINARY –ö–û–î–ò–†–û–í–ê–ù–ò–Ø –ö COUNTRY ===")

# –°–æ–∑–¥–∞–µ–º Binary –∫–æ–¥–∏—Ä–æ–≤—â–∏–∫ –¥–ª—è country
binary_encoder = ce.BinaryEncoder(cols=['country'])

# –ü—Ä–∏–º–µ–Ω—è–µ–º –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ
country_binary_encoded = binary_encoder.fit_transform(wine_df[['country']])

print(f"–†–∞–∑–º–µ—Ä –∏—Å—Ö–æ–¥–Ω–æ–≥–æ DataFrame: {wine_df.shape}")
print(f"–†–∞–∑–º–µ—Ä –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: {country_binary_encoded.shape}")

# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –Ω–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
binary_columns = [col for col in country_binary_encoded.columns if col.startswith('country_')]
print(f"\n‚úÖ –°–æ–∑–¥–∞–Ω–æ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {len(binary_columns)}")
print(f"–ù–∞–∑–≤–∞–Ω–∏—è –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {binary_columns}")

# –î–æ–±–∞–≤–ª—è–µ–º –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∫ –∏—Å—Ö–æ–¥–Ω–æ–º—É DataFrame
wine_df_with_binary = pd.concat([wine_df, country_binary_encoded], axis=1)

print(f"\n–†–∞–∑–º–µ—Ä DataFrame –ø–æ—Å–ª–µ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è Binary –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è: {wine_df_with_binary.shape}")
print(f"–î–æ–±–∞–≤–ª–µ–Ω–æ —Å—Ç–æ–ª–±—Ü–æ–≤: {wine_df_with_binary.shape[1] - wine_df.shape[1]}")
```

```python
# –ê–Ω–∞–ª–∏–∑ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ Binary –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è –Ω–∞ —Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
print("=== –ê–ù–ê–õ–ò–ó –≠–§–§–ï–ö–¢–ò–í–ù–û–°–¢–ò BINARY –ö–û–î–ò–†–û–í–ê–ù–ò–Ø ===")

unique_countries_count = wine_df['country'].nunique()
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å—Ç—Ä–∞–Ω: {unique_countries_count}")
print(f"Binary –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ —Å–æ–∑–¥–∞–ª–æ: {len(binary_columns)} –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")
print(f"OneHot –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ —Å–æ–∑–¥–∞–ª–æ –±—ã: {unique_countries_count} –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")

efficiency = unique_countries_count / len(binary_columns)
print(f"üíæ –≠–∫–æ–Ω–æ–º–∏—è –º–µ—Å—Ç–∞: –≤ {efficiency:.1f} —Ä–∞–∑–∞ –º–µ–Ω—å—à–µ —Å—Ç–æ–ª–±—Ü–æ–≤!")

# –§–æ—Ä–º—É–ª–∞ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –±–∏—Ç–æ–≤
required_bits = math.ceil(math.log2(unique_countries_count))
print(f"\nüìä –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ —Ç—Ä–µ–±—É–µ—Ç—Å—è –±–∏—Ç–æ–≤: ‚åàlog‚ÇÇ({unique_countries_count})‚åâ = {required_bits}")
print(f"üìä –§–∞–∫—Ç–∏—á–µ—Å–∫–∏ —Å–æ–∑–¥–∞–Ω–æ: {len(binary_columns)} –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")

print(f"\n=== –ü–†–ò–ú–ï–†–´ BINARY –ö–û–î–ò–†–û–í–ê–ù–ò–Ø ===")
sample_countries = ['US', 'Italy', 'France', 'Spain', 'Portugal']
for country in sample_countries:
    if country in wine_df['country'].values:
        idx = wine_df[wine_df['country'] == country].index[0]
        country_row = country_binary_encoded.loc[idx]
        
        print(f"–°—Ç—Ä–∞–Ω–∞: {country}")
        binary_values = []
        for col in binary_columns:
            binary_values.append(str(country_row[col]))
        print(f"Binary –∫–æ–¥: [{', '.join(binary_values)}]")
        print("-" * 30)

print(f"\nüéØ –ò–¢–û–ì–û–í–´–ô –û–¢–í–ï–¢:")
print(f"–ü–æ—Å–ª–µ –¥–≤–æ–∏—á–Ω–æ–≥–æ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–∞ 'country' –æ–±—Ä–∞–∑–æ–≤–∞–ª–æ—Å—å {len(binary_columns)} –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")
```
